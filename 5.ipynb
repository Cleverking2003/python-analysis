```python
import scipy.integrate as intg
import scipy.interpolate as interp
import scipy.stats as st
from numpy import sin
import numpy as np
import sympy
import matplotlib.pyplot as plt
```

# 1

```python
print(intg.quad(lambda x : 4 * sin(x) - 3, 0, 5))
print(intg.dblquad(lambda x, y : x ** 2 + y ** 2, 0, 1, 0, 1))
```

# 2

```python
def f(y, x):
    return x / 5 # Правая часть (ПЧ) дифференциального уравнения 

y0 = 1.0  # Начальное условие
k_step = 200  # Количество шагов разбиения отрезка для численного интегрирования
x_output = np.linspace(0, 2, k_step)  
y_result = intg.odeint(f, y0, x_output)
y_result = y_result[:, 0]
print(y_result[-1])
```

# 3 

```python
xs = np.linspace(0, 20, 20, endpoint=True)
ys = (- np.sin(xs)) ** 2

xp = np.linspace(0, 20, 100, endpoint=True)
fun_l = interp.interp1d(xs, ys)  # Вызываем линейную интерполяцию
fun_c = interp.interp1d(xs, ys, kind='cubic')  # Вызываем кубическую интерполяцию

plt.plot(xs, ys, 'o', xp, fun_l(xp), '-', xp, fun_c(xp), '--')
```

# 4

```python
temps = st.uniform.rvs(loc = 27, scale = 30 - 27, size = 20)
vals, counts = np.unique(temps, return_counts=True)

dist = st.rv_discrete(values=(vals, counts / 20))
print(dist.cdf(29))
print(dist.pmf(29))


fig = plt.figure()
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)
ax1.plot(vals, dist.cdf(vals))
ax2.plot(vals, dist.pmf(vals))
ax1.set_title('cdf')
ax2.set_title('pmf')
```
