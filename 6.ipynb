```python
import pandas as pd
import numpy as np
```

```python
def series_data(name, index):
    nums = np.arange(10000)
    nums = nums[nums > 0]
    nums = nums[nums % 2 != 0]
    nums = nums[nums % 15 == 0]
    s = pd.Series(nums, pd.RangeIndex(1, len(nums) + 1), name = name)
    for i in range(index, len(s) + 1):
        if i % 2 == 0:
            s[i] -= 70
        else:
            s[i] *= 2
    return max(s[index], s[index + 1], s[index - 5])

print(series_data('Название серии', 8))
print(series_data('Название серии', 18))
```

```python
def frame_data(values, indices):
    df = pd.DataFrame(values, indices)
    df.sort_index(ascending=False, inplace=True)
    return df.iloc[1, -1]

print(frame_data([[1,11,21], [1,2,3], [-1,-2,-3]], ['V','E','F']))
print(frame_data([[111,111,31,41,51,116], [141,15,61,6,717,160], [77,82,91,324,314,10]], ['Z','E','F']))
```
